```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rgb_1 = np.load('./visualization_imgs/rgb-bathroom.npy')
rgb_1 = np.transpose(rgb_1, (1,2,0))

rgb_2 = np.load('./visualization_imgs/rgb_dining_room-rgb_01383).npy')
rgb_2 = np.transpose(rgb_2, (1,2,0))

rgb_3 = np.load('./visualization_imgs/rgb_dining_room-rgb_01448.npy')
rgb_3 = np.transpose(rgb_3, (1,2,0))

rgb_5 = np.load('./visualization_imgs/rgb_home_office-rgb_00394).npy')
rgb_5 = np.transpose(rgb_5, (1,2,0))

rgb_4 = np.load('./visualization_imgs/rgb_living_room-rgb_01286.npy')
rgb_4 = np.transpose(rgb_4, (1,2,0))


depth_our_1 = np.load('./visualization_imgs/ours/depth_bathroom.npy')
depth_our_2 = np.load('./visualization_imgs/ours/depth_dining_room-rgb_01383.npy')
depth_our_3 = np.load('./visualization_imgs/ours/depth_dining_room-rgb_01448.npy')
depth_our_4 = np.load('./visualization_imgs/ours/depth_dliving_room-rgb_01286.npy')
depth_our_5 = np.load('./visualization_imgs/ours/depth_home_office-rgb_00394.npy')


depth_adabin_1 = np.load('./visualization_imgs/adabin/adabin_rgb_00675.npy')
depth_adabin_2 = np.load('./visualization_imgs/adabin/adabin_rgb_01383.npy')
depth_adabin_3 = np.load('./visualization_imgs/adabin/adabin_rgb_01448.npy')
depth_adabin_4 = np.load('./visualization_imgs/adabin/adabin_rgb_01286.npy')
depth_adabin_5 = np.load('./visualization_imgs/adabin/adabin_rgb_00394.npy')


depth_zoe_1 = np.load('./visualization_imgs/zoedepth/zodepth_rgb_00675.npy')
depth_zoe_2 = np.load('./visualization_imgs/zoedepth/zodepth_rgb_01383.npy')
depth_zoe_3 = np.load('./visualization_imgs/zoedepth/zodepth_rgb_01448.npy')
depth_zoe_4 = np.load('./visualization_imgs/zoedepth/zodepth_rgb_01286.npy')
depth_zoe_5 = np.load('./visualization_imgs/zoedepth/zodepth_rgb_00394.npy')


gt_1 = np.load('./visualization_imgs/gt_tensor20.npy')
gt_2 = np.load('./visualization_imgs/gt_tensor300.npy')
gt_3 = np.load('./visualization_imgs/gt_tensor340.npy')
gt_4 = np.load('./visualization_imgs/gt_tensor550.npy')
gt_5 = np.load('./visualization_imgs/gt_tensor360.npy')
```

```python
fig, axs = plt.subplots(5, 5, figsize=(10, 8))  # Create a grid of 5x5 subplots
# Titles for the first row
titles = ['RGB Image', 'Ours', 'AdaBin', 'ZoeDepth', 'GT']
# Manually set each subplot


axs[0, 0].imshow(rgb_1)
axs[0, 1].imshow(depth_our_1, cmap='plasma', vmin=depth_our_1.min(), vmax=depth_our_1.max())
axs[0, 2].imshow(depth_adabin_1, cmap='plasma', vmin=depth_adabin_1.min(), vmax=depth_adabin_1.max())
axs[0, 3].imshow(depth_zoe_1, cmap='plasma', vmin=depth_our_1.min(), vmax=depth_our_1.max())
axs[0, 4].imshow(gt_1, cmap='plasma', vmin=gt_1.min(), vmax=gt_1.max())

axs[1, 0].imshow(rgb_2)
axs[1, 1].imshow(depth_our_2, cmap='plasma', vmin=depth_our_2.min(), vmax=depth_our_2.max())
axs[1, 2].imshow(depth_adabin_2, cmap='plasma', vmin=depth_adabin_2.min(), vmax=depth_adabin_2.max())
axs[1, 3].imshow(depth_zoe_2, cmap='plasma', vmin=depth_our_2.min(), vmax=depth_our_2.max())
axs[1, 4].imshow(gt_2, cmap='plasma', vmin=gt_2.min(), vmax=gt_2.max())

axs[2, 0].imshow(rgb_3)
axs[2, 1].imshow(depth_our_3, cmap='plasma', vmin=depth_our_3.min(), vmax=depth_our_3.max())
axs[2, 2].imshow(depth_adabin_3, cmap='plasma', vmin=depth_adabin_3.min(), vmax=depth_adabin_3.max())
axs[2, 3].imshow(depth_zoe_3, cmap='plasma', vmin=depth_our_3.min(), vmax=depth_our_3.max())
axs[2, 4].imshow(gt_3, cmap='plasma', vmin=gt_3.min(), vmax=gt_3.max())

axs[3, 0].imshow(rgb_4)
axs[3, 1].imshow(depth_our_4, cmap='plasma', vmin=depth_our_4.min(), vmax=depth_our_4.max())
axs[3, 2].imshow(depth_adabin_4, cmap='plasma', vmin=depth_adabin_4.min(), vmax=depth_adabin_4.max())
axs[3, 3].imshow(depth_zoe_4, cmap='plasma', vmin=depth_our_4.min(), vmax=depth_our_4.max())
axs[3, 4].imshow(gt_4, cmap='plasma', vmin=gt_4.min(), vmax=gt_4.max())

axs[4, 0].imshow(rgb_5)
axs[4, 1].imshow(depth_our_5, cmap='plasma', vmin=depth_our_5.min(), vmax=depth_our_5.max())
axs[4, 2].imshow(depth_adabin_5, cmap='plasma', vmin=depth_adabin_5.min(), vmax=depth_adabin_5.max())
axs[4, 3].imshow(depth_zoe_5, cmap='plasma', vmin=depth_our_5.min(), vmax=depth_our_5.max())
axs[4, 4].imshow(gt_5, cmap='plasma', vmin=gt_5.min(), vmax=gt_5.max())

for i in range(5):
    axs[0, i].set_title(titles[i])
for ax in axs.flat:
    ax.axis('off')


plt.tight_layout()
# Adjust subplot parameters to reduce gaps.
plt.subplots_adjust(wspace=0.05, hspace=0)


plt.savefig('visualization.png', dpi=100, bbox_inches='tight')
```
